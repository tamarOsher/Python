כדי להפעיל פונקציה על עמודה משתמשים במתודה apply:



```python
def get_company(s):
  if s.find('Mazda RX4') > -1:
    return 'Mazda'
  elif s.find('Audi') > -1:
    return 'Audi'
  else:
    return 'Unknown'
```

```python
df['company'] = df.model.apply(get_company)
```

```python
df.head()
```
